```
import pandas as pd
df=pd.read_csv('/content/Iris.csv')
df
```

# **USING LOGISITIC REGRESSION**

```
X=df.drop(['Id','Species'],axis='columns')
X=X.values
X
```

```
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```
dfle=df
dfle['Species']=le.fit_transform(dfle['Species'])
dfle
```

```
import numpy as np
y=dfle['Species'].values
y1=np.reshape(y,(-1,1))   #Reshaping the array to make it a feature
```

```
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
y=ohe.fit_transform(y1).toarray()
y1
```

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y1,test_size=0.2)
```

```
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```
type(X_train)
```

# **VERY IMPORTANT**

TARGET SHAPE SHOULD BE A (X,1) Array not anything else. If any other input shape is given it gives bad input shape

```
model.fit(X_train,y_train)
```

```
model.fit(X_train,y_train)
```

```
score=model.score(X_test,y_test)
score
```

```
model.predict([[4.5,3.4,1.5,0.3]])   #0 means iris setosa. 1 means iris versicolor.2 means iris virginica
```

```
y_pred=model.predict(X_test)         #To test the accuracy of the model even more better we use a confusion_matrix
from sklearn.metrics import confusion_matrix

cm=confusion_matrix(y_test,y_pred)  
cm
```

```
import seaborn as sn
import matplotlib.pyplot as plt
plt.figure(figsize=(15,7))
sn.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
